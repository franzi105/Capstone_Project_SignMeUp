```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from tensorflow import keras
from tensorflow.keras import layers


# Load the data
X = np.load("../data/X-data.npy")
y = np.load("../data/y-data.npy")

# Reshape y into a 1D array
y = np.argmax(y, axis=1)

#defining signs --> edit for specific subset of data
actions = np.array ( ['alligator', 'radio', 'moon', 'sleep', 'grandpa', 'tiger', 'pencil', 'sleepy', 'grandma', 'chocolate'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)


# Train and evaluate the Naive Bayes classifier
nb_clf = GaussianNB()
nb_clf.fit(X_train.reshape(X_train.shape[0], -1), y_train)
nb_pred = nb_clf.predict(X_test.reshape(X_test.shape[0], -1))
nb_acc = accuracy_score(y_test, nb_pred)

# Train and evaluate the Decision Tree classifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(X_train.reshape(X_train.shape[0], -1), y_train)
dt_pred = dt_clf.predict(X_test.reshape(X_test.shape[0], -1))
dt_acc = accuracy_score(y_test, dt_pred)

# Train and evaluate the Random Forest classifier
rf_clf = RandomForestClassifier()
rf_clf.fit(X_train.reshape(X_train.shape[0], -1), y_train)
rf_pred = rf_clf.predict(X_test.reshape(X_test.shape[0], -1))
rf_acc = accuracy_score(y_test, rf_pred)

# Train and evaluate the Logistic Regression classifier
lr_clf = LogisticRegression()
lr_clf.fit(X_train.reshape(X_train.shape[0], -1), y_train)
lr_pred = lr_clf.predict(X_test.reshape(X_test.shape[0], -1))
lr_acc = accuracy_score(y_test, lr_pred)

# Train and evaluate the K-Nearest Neighbors classifier
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train.reshape(X_train.shape[0], -1), y_train)
knn_pred = knn_clf.predict(X_test.reshape(X_test.shape[0], -1))
knn_acc = accuracy_score(y_test, knn_pred)


# Train and evaluate the Support Vector Machine classifier
svm_clf = SVC()
svm_clf.fit(X_train.reshape(X_train.shape[0], -1), y_train)
svm_pred = svm_clf.predict(X_test.reshape(X_test.shape[0], -1))
svm_acc = accuracy_score(y_test, svm_pred)

num_classes = len(np.unique(y_train))

# Create a Sequential model for the CNN classifier
cnn_clf = keras.Sequential(
    [
        layers.Conv2D(32, kernel_size=(3, 3), activation="relu", input_shape=(X_train.shape[1], X_train.shape[2], 1)),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Conv2D(64, kernel_size=(3, 3), activation="relu"),
        layers.MaxPooling2D(pool_size=(2, 2)),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation="softmax"),
    ]
)

# Compile the CNN classifier
cnn_clf.compile(optimizer="adam", loss=keras.losses.SparseCategoricalCrossentropy())
# Train the CNN classifier
cnn_clf.fit(X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2], 1), y_train, epochs=100, batch_size=25)

# Evaluate the CNN classifier
cnn_pred = cnn_clf.predict(X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2], 1))
cnn_pred = np.argmax(cnn_pred, axis=1)
cnn_acc = accuracy_score(y_test, cnn_pred)

```

```python
print("Naive Bayes Classifier Accuracy:", nb_acc)
print("Decision Tree Classifier Accuracy:", dt_acc)
print("Random Forest Classifier Accuracy:", rf_acc)
print("Logistic Regression Classifier Accuracy:", lr_acc)
print("Support Vector Machine Classifier Accuracy:", svm_acc)
print("CNN Classifier Accuracy:", cnn_acc)
```
